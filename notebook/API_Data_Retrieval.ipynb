# API Data Retrieval Using OpenAQ API

```python
from openaq import OpenAQ
import pandas as pd 
from dotenv import load_dotenv
import os
```

```python
load_dotenv()

api = os.getenv('OPEN_AQ_API')
```

```python
client = OpenAQ(api_key=api)
```

```python
#Testing The Api
Indian_AQI_Retrival = client.locations.list(
    
    bbox=(74.955323,26.019799,79.393799,29.821974),
    limit=10
)
```

```python
Indian_AQI_Retrival.meta
```

```python
Indian_AQI_Retrival.results[0].__dict__
```

```python
Indian_AQI_Retrival2 = client.locations.list(
    countries_id=9,
    limit=1000
)
```

```python
Indian_AQI_Retrival2.meta
```

```python
Indian_AQI_Retrival2.results
```

```python
Indian_AQI_Retrival2.meta
```

```python
location_info = {}

for i in Indian_AQI_Retrival2.results:
    location_info[i.id] = i.name
```

```python
location_info
```

```python
len(location_info)
```

```python
location_df = pd.DataFrame(list(location_info.items()), columns=['ID', 'Location'])
```

```python
location_df.to_csv('output.csv',index=False)
```

**This Code Would Create a Location_Info File Which Contains All The Location Info**

```python
data = []
for i in Indian_AQI_Retrival2.results:
    instruments = list({inst.name for inst in i.instruments})  # unique instrument names
    instrument_name = instruments[0] if instruments else None  # pick any one
    data.append({
        'location_id': i.id,
        'location_name': i.name,
        'country_name': i.country.name,
        'provider_name': i.provider.name,
        'owner_name': i.owner.name,
        'is_mobile': i.is_mobile,
        'is_monitor': i.is_monitor,
        'instrument_name': instrument_name
    })

df = pd.DataFrame(data)
df.to_csv('locations_info1.csv', index=False)
```
