# File Compressions

```python
import pandas as pd
```

```python
aqidf = pd.read_csv('../pipeline/data/AQI_Data_2024_2025.csv')
```

```python
aqidf.sample(10)
```

```python
combined_df = aqidf
```

**Compressing The File**

```python
combined_df['date'] = combined_df['datetime'].dt.date; new_df = combined_df.groupby(['date', 'location_id', 'sensors_id', 'location', 'lat', 'lon', 'parameter', 'units'])['value'].mean().reset_index()
```

**Transforming 20 Million Rows in 273K Files**

```python
new_df.to_csv('Compressed_AQI.csv',index=False)
```
