```python
import pandas as pd
```

```python
df = pd.read_csv('pipeline\data\Compressed_AQI_Location_Info.csv')
```

```python
df['parameter'].unique()
```

```python
df_cleaning = df = df[~((df['parameter'] == 'pm25') & ((df['value'] > 500) | (df['value'] < 0)))]
```

```python
df_cleaning = df_cleaning[~((df_cleaning['parameter'] == 'temperature') & ((df['value'] < -10) | (df['value'] > 55)))]
```

```python
# Removing negative values and values above 300 µg/m³ for 'o3'
df_cleaning = df_cleaning[~((df_cleaning['parameter'] == 'o3') & ((df_cleaning['value'] < 0) | (df_cleaning['value'] > 300)))]
```

```python
# Removing negative values and values above 1000 ppb for 'so2'
df_cleaning = df_cleaning[~((df_cleaning['parameter'] == 'so2') & ((df_cleaning['value'] < 0) | (df_cleaning['value'] > 1000)))]
```

```python
df_cleaning[df_cleaning['parameter']=='so2'].sort_values('value')
```

```python
df_cleaning = df_cleaning[~((df_cleaning['parameter'] == 'co') & ((df_cleaning['value'] < 0) | (df_cleaning['value'] > 100)))]
```

```python
df_cleaning[df_cleaning['parameter']=='no2'].sort_values('value')
```

```python
df_cleaning = df_cleaning[~((df_cleaning['parameter'] == 'relativehumidity') & ((df_cleaning['value'] < 0) | (df_cleaning['value'] > 100)))]
```

```python
df_cleaning = df_cleaning[~((df_cleaning['parameter'] == 'pm10') & ((df_cleaning['value'] < 0) | (df_cleaning['value'] > 100)))]
```

```python
df_cleaning = df_cleaning[df_cleaning['parameter'] != 'um003']
```

```python
# Removing rows where 'no' has 0 or extremely high values
df_cleaning = df_cleaning[~((df_cleaning['parameter'] == 'no') & ((df_cleaning['value'] == 0) | (df['value'] > 1e6)))]
```

```python
df_cleaning = df_cleaning[~((df_cleaning['parameter'] == 'no') & (df_cleaning['value'] > 1000))]
```

```python
df_cleaning = df_cleaning[~((df_cleaning['parameter'] == 'no2') & ((df_cleaning['value'] < 0) | (df['value'] > 200)))]
```
